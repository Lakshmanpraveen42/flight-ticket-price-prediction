```python
def replace(a):
    if a.startswith("UK"):
        return a.replace("UK-","1")
    if a.startswith("I5"):
        return a.replace("I5-","2")
    if a.startswith("AI"):
        return a.replace("AI-","3")
    if a.startswith("G8"):
        return a.replace("G8-","4")
    if a.startswith("SG"):
        return a.replace("SG-","5")
    if a.startswith("6E"):
        return a.replace("6E-","6")    
```

```python
x_train['flight']=x_train['flight'].apply(replace)
x_test['flight']=x_test['flight'].apply(replace)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for col in ["airline","source_city","departure_time","stops","arrival_time","destination_city"]:
    x_train[col]=encoder.fit_transform(x_train[col])
    x_test[col]=encoder.fit_transform(x_test[col])
    mapping=dict((zip(encoder.classes_,encoder.transform(encoder.classes_))))
    print(f"{col} : ",mapping)
```
